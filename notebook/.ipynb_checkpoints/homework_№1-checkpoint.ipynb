#### Выбрать тематику данных: лучшего всего для этого посмотреть, какие сервисы есть в публичном доступе там, откуда вы сможете эти данные собрать. Нам нужны RESTful- и SOAP-сервисы.

```python
import requests
import json


def view_dict(dictionary: dict, view_text = str()):
    for key, value in dictionary.items():
        view_text += f'{key}: {value};\n'
    return view_text
          
def write_result_to_file(text: str, file_name: str, param = {}):
    with open(file_name, 'w') as file:
        file_text = file.write(text)
    return file_text

headers = {'X-Yandex-API-Key':'a2891b29-7e70-4943-93ba-a44a0676a107'}
get = requests.get('https://api.weather.yandex.ru/v1/forecast?', headers=headers)

status_code = get.status_code
response_headers = view_dict(get.headers)
response_text = view_dict(json.loads(get.text))

status_code_text = f"""
=======================================================\n
RESPONSE CODE:\n
{status_code}\n
=======================================================\n
RESPONSE HEADERS:\n
{response_headers}
=======================================================\n
RESPONSE CONTEXT:\n
{response_text}
"""
write_result_to_file(status_code_text, 'data/response_data.txt')

print(f'{status_code_text}')

```
